```python
# importing the client keys
cid = config.cid
secret = config.secret

# setting up the spotify object
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
# getting tracks audio features
acousticness = []
danceability = []
energy = []
key = []
instrumentalness = []
liveness = []
loudness = []
speechiness = []
tempo = []
valence = []
duration_ms = []

for i in range(0,1000,50):

    features = sp.audio_features(track_id[i:i+50])
    
    for i, t in enumerate(features):

        try:
            acousticness.append(t['acousticness'])
        except:
            acousticness.append('')
            
        try:
            danceability.append(t['danceability'])
        except:
            danceability.append('')
        
        try:
            energy.append(t['energy'])
        except:
            energy.append('')
            
        try:
            key.append(t['key'])
        except:
            key.append('')
            
        try:
            instrumentalness.append(t['instrumentalness'])
        except:
            instrumentalness.append('')
            
        try:
            liveness.append(t['liveness'])
        except:
            liveness.append('')
            
        try:
            loudness.append(t['loudness'])
        except:
            loudness.append('')
            
        try:
            speechiness.append(t['speechiness'])
        except:
            speechiness.append('')
               
        try:
            tempo.append(t['tempo'])
        except:
            tempo.append('')
            
        try:
            valence.append(t['valence'])
        except:
            valence.append('')
            
        try:
            duration_ms.append(t['duration_ms'])
        except:
            duration_ms.append('')
        
# dictionary of audio_features 
audio_features_dict = {'track_id': track_id, 'acousticness': acousticness,
                       'danceability': danceability, 'energy': energy,
                       'key': key, 'instrumentalness': instrumentalness,
                       'liveness': liveness, 'loudness': loudness,
                       'speechiness': speechiness, 'tempo': tempo,
                       'valence': valence, 'duration_ms': duration_ms
                      } 

    
audio_features_df = pd.DataFrame(audio_features_dict)
    
audio_features_df.head()     
```
